## Calculate Active Return ($\alpha$) and Volatility (W.R.T S&P 500, $\beta$)

```python
def get_metrics(group) -> pd.Series:
    if len(group) < 502:
        return pd.Series({"Beta": np.nan, "Alpha": np.nan, "Volatility": np.nan, "Annual Return": np.nan})

    volatility = group["Daily Return"].std() * (251 ** 0.5)
    
    # find beta
    covariance_matrix = np.cov(group["Daily Return"], group["Index Return"])
    covariance = covariance_matrix[0, 1]
    variance = group["Index Return"].var()
    beta = covariance / variance

    # find alpha
    # risk free rate is about 4.2%
    # src: https://www.cnbc.com/quotes/US10Y
    rfr = 0.042

    annual_stock_return = group["Daily Return"].mean() * 251
    annual_market_return = group["Index Return"].mean() * 251
    expected_return = rfr + beta * (annual_market_return - rfr)
    alpha = annual_stock_return - expected_return

    return pd.Series({
        "Beta": beta,
        "Alpha": alpha,
        "Volatility": volatility,
        "Annual Return": annual_stock_return
    })

df_metrics = df_stocks_index.groupby("Symbol").apply(get_metrics, include_groups=False).reset_index()
df_metrics = df_metrics.merge(df_companies[["Symbol", "Shortname", "Sector"]], on="Symbol", how="left")
df_metrics.dropna(inplace=True)


print(f"\n{"=" * 75}")
print("Stocks sorted by volatility (descending):")
print(f"{"=" * 75}")
display(df_metrics.sort_values(by="Volatility", ascending=False))
print(f"\n{"=" * 75}")
print("Stocks sorted by alpha value (descending):")
print(f"{"=" * 75}")
display(df_metrics.sort_values(by="Alpha", ascending=False))
print(f"\n{"=" * 75}")
print("Stocks sorted by beta value (descending):")
print(f"{"=" * 75}")
display(df_metrics.sort_values(by="Beta", ascending=False))

df_metrics.to_csv("sp500_alphabeta.csv", index=False)
```
